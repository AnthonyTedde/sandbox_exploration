```python
import numpy as np
import matplotlib.pyplot as plt
```

# Orthogonal vector

```python
origin = np.array([[0, 0], [0, 0]])

V1 = np.array([1, 1])
V2 = np.array([-2, 2])
```

```python
print(np.dot(V1, V2))
```

```python
V = np.array([V1, V2])
fig = plt.figure()
ax = fig.gca()
ax.quiver(*origin, V[:, 0], V[:, 1],
           color=["tab:red", "tab:blue"],
           angles="xy", 
           scale_units="xy", 
           scale=1)
ax.grid(linestyle="-.")
ax.set_xlim([-3, 3])
ax.set_ylim([-1, 3])
ax.set_xticks(np.arange(-3, 3, step=1))
ax.set_yticks(np.arange(-1, 3, step=1))
ax.set_aspect(aspect="equal")
```
